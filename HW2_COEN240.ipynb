```python
#Q1 Write a python function that takes a sequence of numbers and determines if all the numbers are distinct from each other or if there are any duplicates.
def test_distinct(numbers):
    return len(numbers) != len(set(numbers))

print(test_distinct([1,5,7,9]))
print(test_distinct([2,4,5,5,7,9]))
```

```python
#Q2

import numpy as np

def zero_indices(arr):
    return np.where(arr == 0)
a = np.array([1,0,2,0,3,0,4,5,6,7,8])
print(zero_indices(a))
```

```python
#Q3

import numpy as np
def mean_squared_error(actual, predicted, W0, W1):
    predicted = W0 + (W1 * np.array(predicted))
    mse = np.mean((np.array(actual) - np.array(predicted)) ** 2)
    return mse
input_W0 = 0.25
input_W1 = 0.75
x=[100, 200, 300, 400, 500]
y=[65, 155, 220, 310, 370]

mean_squared_error(x,y,input_W0,input_W1)
```
